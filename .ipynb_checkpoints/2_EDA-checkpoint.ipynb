# Exploratory Data Analysis
In this notebook, I will explore our training data in details using visualisations to gain further insights.
In summary, I found out that there are some abnormal values in our data. Instead of dropping the rows/column entirely, I will be replacing it with the column's median value. I have also found out that Utilization Of Unsecured Lines has the most predictive power in this dataset.

```python
# Base
import pandas as pd
import numpy as np

# Visualisation
import seaborn as sns
import matplotlib.pyplot as plt

%config InlineBackend.figure_format = 'retina'
%matplotlib inline
```

```python
training = pd.read_pickle('datasets/clean_training.pickle')
test = pd.read_pickle('datasets/clean_test.pickle')
data_dict = pd.read_pickle('datadict/data_dict.pickle')
```

```python
data_dict
```

```python
sns.pairplot(training,hue='dlq')  #First look
```

```python
training.columns
```

## Utilization Of Unsecured Lines (Util_ratio)
Total balance on credit cards and personal lines of credit except real estate and no installment debt like car loans divided by the sum of credit limits

```python
ax = sns.boxplot(training['util_ratio'])  # Many outliers!
ax.set_title('Boxplot')
```

```python
training['util_ratio'].describe()
```

```python
training[training['util_ratio'] > 1]['util_ratio'].count()
```

```python
median_util_ratio = training[training['util_ratio'] > 1]['util_ratio'].median()
```

There seems to be some abnormal entries in the data as the value should be between 0 and 1. There are 3321 instances where ratio > 1. We will replace them with median.

```python
# replace median if value > 1
training['util_ratio'] = training['util_ratio'].map(lambda x: median_util_ratio if x > 1 else x) 
```

### Test set

```python
ax = sns.boxplot(test['util_ratio'])
ax.set_title('Boxplot')
```

We will replace values in the test set that are more than 1 with the median as well.

```python
# replace median if value > 1
test['util_ratio'] = test['util_ratio'].map(lambda x: median_util_ratio if x > 1 else x) 
```

Monthly debt payments, alimony,living costs divided by monthy gross income

It seems abnormal to have debt ratio more than 10,000. It means that the monthly debt repayment is 10,000 times per dollar of gross income. We will treat them as abnormal data by replacing with median. 

```python
median_debtratio = training[training['debtratio']<10000]['debtratio'].median()
```

```python
# replace median if value > 10000
training['debtratio'] = training['debtratio'].map(lambda x: median_debtratio if x > 10000 else x)
```
